```python
from bs4 import BeautifulSoup
import urllib.request
import pickle
import time
import re
def get_text(URL):
    source_code_from_URL = urllib.request.urlopen(URL)
    soup = BeautifulSoup(source_code_from_URL, 'lxml', from_encoding='utf-8')
    text = ''
    for item in soup.find_all('div', id='articleBodyContents'):
        text = text + str(item)
        
    return text

# # Daum
# def get_text(URL):
#     source_code_from_URL = urllib.request.urlopen(URL)
#     soup = BeautifulSoup(source_code_from_URL, 'lxml', from_encoding='utf-8')
#     text = ''
#     for item in soup.find('div', id='harmonyContainer').find_all('p'):
#         text = text + str(item)
        
#     return text
```

```python
# 0(정치)
with open('정치추가.pickle', 'rb') as f:
    data = pickle.load(f)

index = 200
for i in range(0,200):
    URL =  data[i]
    OUTPUT_FILE_NAME = ""
    
    if index<1000:
        OUTPUT_FILE_NAME = 'newsdata/0/'+ '+00'+ str(index) + 'NewsData.txt'
    else :
        OUTPUT_FILE_NAME = 'newsdata/0/'+ '+0'+ str(index) + 'NewsData.txt'
    
    index = index +1
    
    text = get_text(URL)
    clean = text.replace("\n","")
    clean = clean.replace("\t","")
    clean = clean.replace("\xa0"," ")
    clean = re.sub('<script.*?/script>','',clean)
    clean = re.sub('<img.*?/img>','',clean)
    clean = re.sub('<table.*?/table>','',clean)
    clean = re.sub('<.*?>','',clean)
    clean = re.sub('&...','',clean)
    r = clean.rfind("@")
    clean = clean[:r]
    r = clean.rfind(".")
    clean = clean[:r]
    write_file = open(OUTPUT_FILE_NAME, 'w',encoding='UTF-8')
    write_file.write(clean)
    write_file.close()
    time.sleep(2.5)
    
```
